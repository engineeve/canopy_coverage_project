# Patch-based Architectures

The following notebook provides development of two established CNN architectures (LeNet and VGG16). Both models are trained useing the ISPRS dataset of the city of Vaihingen in Germany: (http://www2.isprs.org/commissions/comm3/wg4/2d-sem-label-vaihingen.html).


Import libraries

```python
import numpy as np
from skimage import io
from tqdm import tqdm_notebook as tqdm
from sklearn.metrics import confusion_matrix
import random
import itertools
# Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import os
import cv2

from matplotlib import ticker
import seaborn as sns
from PIL import Image
from sklearn.feature_extraction import image
import glob

from sklearn.model_selection import train_test_split
from sklearn import metrics
from imblearn.under_sampling import RandomUnderSampler
from sklearn.utils import shuffle

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
from keras.layers import Input, Dropout, Flatten, Conv2D, MaxPooling2D, Dense, Activation
from keras.layers.normalization import BatchNormalization
from keras.layers.core import Activation
from keras.optimizers import RMSprop, adam
from keras.utils import np_utils
from keras import backend as K
import pickle

from datetime import date
```

Load Global Variables

```python
CHANNELS = 3 # Number of input channels (e.g. RGB)
IMG_PATH = '/home/niamh/notebooks/PATCH_BASED/patch_32_stride_18/train_patches/total_train/'
SAMPLE_RATE = 1
```

# 1. LeNet Model

## Data Pre-processing

```python
ROWS = 33
COLS = 33

image_filenames = [IMG_PATH+i for i in os.listdir(IMG_PATH) if not i.startswith('.')]  
image_filenames= random.sample(image_filenames, int(len(image_filenames)*SAMPLE_RATE))

count = len(image_filenames)
train_data = np.ndarray((count , ROWS, COLS, CHANNELS), dtype=np.float)

# Iterate throuigh the filenames and for each one load the image, resize and normalise
for i, image_file in enumerate(image_filenames): 
    image = cv2.imread(image_file)
    image = cv2.resize(image, (ROWS, COLS), interpolation=cv2.INTER_CUBIC)  
    train_data[i] = image
    train_data[i] = train_data[i]/255
    if i%1000 == 0: print('Processed {} of {}'.format(i, count))
        
print("Train shape: {}".format(train_data.shape))  
```

```python
# Extract the lables from the last characters in the training filename
train_labels = []

for i in image_filenames:
    l = i[-5]
    if l == '3':
        # Tree
        l = '1'
        train_labels.append(int(l))
    else:
        # Non Tree
        l = '0'
        train_labels.append(int(l))
        
# Count the number of clases
num_classes = len(set(train_labels))
num_of_examples = len(train_labels)

# Plot a bar plot of the label distrib
sns.countplot(train_labels)

unique, counts = np.unique(train_labels, return_counts=True)
count_labels = dict(zip(unique, counts))


f = open('training_data_info.txt','w')
f.write('Balance of tree/ non-tree')
f.write(str(count_labels))
f.write(' ============================================ ')
f.write('lenght train_labels > ')
f.write(str(num_of_examples))
f.close()
```

```python
n = 5
plt.figure(figsize=(20, 10)) 

for i in range(n):
    i = i+1
    idx = random.randint(0, len(train_data))
    
    # display original
    ax = plt.subplot(2, n, i)
    plt.imshow(train_data[idx].reshape(ROWS, COLS,3))
    plt.gray()
#     ax.get_xaxis().set_visible(False)
#     ax.get_yaxis().set_visible(False)

    if train_labels[idx] == 1:
        label = 'Tree'
    if train_labels[idx] == 0:
        label = 'Non tree'
    plt.title('Image example '+ str(idx) + ': '+label)
    
plt.show()
```

```python
# convert to binary encoded labels
train_labels_wide = keras.utils.to_categorical(train_labels, num_classes)
```

## Build and compile LeNet Model

Model Parameters

```python
# Model Params
loss='binary_crossentropy'
optimizer=adam(lr=1e-6)

#Training Params
epochs = 10
batch_size = 10
best_weights_filepath = 'LeNet_best_weights.hdf5'
validation_split=0.25
```

```python
# 
activation="relu"

lenet_model = Sequential()
inputShape = (33, 33, 3)
lenet_model.add(Conv2D(20, 5, padding="same",input_shape=inputShape))
lenet_model.add(Activation(activation))
lenet_model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

lenet_model.add(Conv2D(50, 5, padding="same"))
lenet_model.add(Activation(activation))
lenet_model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

lenet_model.add(Flatten())
lenet_model.add(Dense(500))
lenet_model.add(Activation(activation))
 
lenet_model.add(Dense(2))
lenet_model.add(Activation("softmax"))
lenet_model.summary()
```

Compile and fit LeNet Model

```python
lenet_model.compile(loss=loss, optimizer=optimizer, metrics=['accuracy'])

mcp = ModelCheckpoint(best_weights_filepath, monitor="val_loss",
                      save_best_only=True, save_weights_only=False)


history = lenet_model.fit(train_data, 
                          train_labels_wide, 
                          batch_size=batch_size, 
                          epochs=epochs,
                          validation_split=validation_split,
                          verbose=1, shuffle=True,
                          callbacks=[mcp])
```

Evaluate lenet Model during training

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

acc = history.history['acc']
val_acc = history.history['val_acc']

loss_fig = plt.figure()

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(loss, 'blue', label='Training Loss')
plt.plot(val_loss, 'green', label='Validation Loss')
plt.xticks(range(0,epochs)[0::2])
plt.legend()
plt.show()
loss_fig.savefig('LeNet_loss.png')


acc_fig = plt.figure()

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.plot(acc, 'blue', label='Training Accuracy')
plt.plot(val_acc, 'green', label='Validation Accuracy')
plt.xticks(range(0,epochs)[0::2])
plt.legend()
plt.show()

acc_fig.savefig('LeNet_acc.png')

```

```python
# Load best weights form training
lenet_model.load_weights(best_weights_filepath)

# Save final model
filepath = "LeNet_model.mod"
lenet_model.save(filepath)
```

# 2. VGG16 Model

## Data Pre-processing

```python
ROWS = 48
COLS = 48

train_data = np.ndarray((count , ROWS, COLS, CHANNELS), dtype=np.float)

# Iterate throuigh the filenames and for each one load the image, resize and normalise
for i, image_file in enumerate(image_filenames): 
    image = cv2.imread(image_file)
    image = cv2.resize(image, (ROWS, COLS), interpolation=cv2.INTER_CUBIC)  
    train_data[i] = image
    train_data[i] = train_data[i]/255
    if i%1000 == 0: print('Processed {} of {}'.format(i, count))
        
print("Train shape: {}".format(train_data.shape))  
```

```python
train_labels = []

for i in image_filenames:
    l = i[-5]
    if l == '3':
        # Tree
        l = '1'
        train_labels.append(int(l))
    else:
        # No Tree
        l = '0'
        train_labels.append(int(l))
        
# Count number of clases
num_classes = len(set(train_labels))
num_of_examples = len(train_labels)

# convert to binary encoded labels
train_labels_wide = keras.utils.to_categorical(train_labels, num_classes)

unique, counts = np.unique(train_labels, return_counts=True)
count_labels = dict(zip(unique, counts))

num_classes = 2
train_labels_wide = keras.utils.to_categorical(train_labels, num_classes)
```

```python
n = 5
plt.figure(figsize=(20, 10)) 

for i in range(n):
    i = i+1
    idx = random.randint(0, len(train_data))
    
    # display original
    ax = plt.subplot(2, n, i)
    plt.imshow(train_data[idx].reshape(ROWS, COLS,3))
    plt.gray()
#     ax.get_xaxis().set_visible(False)
#     ax.get_yaxis().set_visible(False)

    if train_labels[idx] == 1:
        label = 'Tree'
    if train_labels[idx] == 0:
        label = 'Non tree'
    plt.title('Image example '+ str(idx) + ': '+label)
    
plt.show()
```

## Build and compile VGG16 Model

Model and training parameters 

```python
# Model Params
loss='binary_crossentropy'
optimizer=keras.optimizers.SGD(lr=1e-4, momentum=0.9)

#Training Params
epochs = 10
batch_size = 10
best_weights_filepath = 'VGG16_best_weights.hdf5'
validation_split=0.2
```

Build and compile pre-trained model

```python
trained_vgg16_model = keras.applications.VGG16(weights='imagenet', include_top=False, input_shape = train_data[0].shape)
display(trained_vgg16_model.summary())
```

```python

# add final model layers
vgg16_last_layer = trained_vgg16_model.output
x1 = Flatten()(vgg16_last_layer)
x2 = Dense(256, activation='relu')(x1)
x3 = Dropout(0.5)(x2)
final_layer = Dense(num_classes, activation = 'softmax')(x3)

# Combine pre-built layers with top layers
vgg_model = keras.Model(trained_vgg16_model.input, final_layer)

# Freeze first 17 layers while training 
for layer in vgg_model.layers[:17]:
    layer.trainable = False
    
vgg_model.summary()
```

```python
vgg_model.compile(loss=loss,
                   optimizer=optimizer,
                   metrics=['accuracy'])

mcp = ModelCheckpoint(best_weights_filepath, monitor="val_loss",
                      save_best_only=True, save_weights_only=False)

history = vgg_model.fit(train_data, train_labels_wide, 
          batch_size=batch_size,
          epochs=epochs,
          verbose = 1,
          validation_split = 0.2,
          shuffle=True,
          callbacks=[mcp])
```

Display training information

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

acc = history.history['acc']
val_acc = history.history['val_acc']

loss_fig = plt.figure()

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(loss, 'blue', label='Training Loss')
plt.plot(val_loss, 'green', label='Validation Loss')
plt.xticks(range(0,epochs)[0::2])
plt.legend()
plt.show()
loss_fig.savefig('VGG16_loss.png')


acc_fig = plt.figure()

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.plot(acc, 'blue', label='Training Accuracy')
plt.plot(val_acc, 'green', label='Validation Accuracy')
plt.xticks(range(0,epochs)[0::2])
plt.legend()
plt.show()

acc_fig.savefig('VGG16_acc.png')
```

```python
# Load best weights form training
vgg_model.load_weights(best_weights_filepath)

# Save final model
filepath = "VGG16_model.mod"
vgg_model.save(filepath)
```
